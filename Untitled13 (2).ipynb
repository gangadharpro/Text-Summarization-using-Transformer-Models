```python
# 1. Import Required Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert dataset splits into Pandas DataFrames
df_train = pd.read_csv(r'C:\Users\nikes\OneDrive\Documents\PROJECT DSF\cnn_dailymail\train.csv')
df_val = pd.read_csv(r'C:\Users\nikes\OneDrive\Documents\PROJECT DSF\cnn_dailymail\validation.csv')
df_test = pd.read_csv(r'C:\Users\nikes\OneDrive\Documents\PROJECT DSF\cnn_dailymail\test.csv')
```

```python
# Check for missing values
def check_missing_values(df, name):
    print(f"Missing values in {name}:")
    print(df.isnull().sum())

check_missing_values(df_train, "Train")
check_missing_values(df_val, "Validation")
check_missing_values(df_test, "Test")
```

```python
# Display basic information about the dataset
def display_basic_info(df, name):
    print(f"\nBasic Information for {name}:")
    print(df.info())
    print(df.describe())
    print(df.head())

display_basic_info(df_train, "Training Data")
display_basic_info(df_val, "Validation Data")
display_basic_info(df_test, "Test Data")
```

```python
# EDA: Word Count Analysis
import re
def word_count_analysis(df, name):
    article_lengths = df['article'].apply(lambda x: len(re.findall(r'\w+', x)))
    highlight_lengths = df['highlights'].apply(lambda x: len(re.findall(r'\w+', x)))
    print(f"{name} - Article Length Statistics:\n", article_lengths.describe())
    print(f"{name} - Highlights Length Statistics:\n", highlight_lengths.describe())
    plt.figure(figsize=(10,5))
    sns.histplot(article_lengths, bins=50, kde=True, label='Articles')
    sns.histplot(highlight_lengths, bins=50, kde=True, label='Highlights', color='red')
    plt.legend()
    plt.title(f"{name} - Word Count Distribution")
    plt.show()

word_count_analysis(df_train, "Training Data")
word_count_analysis(df_val, "Validation Data")
word_count_analysis(df_test, "Test Data")
```

```python
import nltk
import spacy
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords

# Load spaCy model with disabled components for efficiency
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])  
nlp.max_length = 5_000_000  # Increase max character limit

# Load stopwords
stop_words = set(stopwords.words('english'))

# Optimized function to process text in chunks
def plot_most_frequent_words(text_series, title, num_words=20):
    text_list = text_series[:1000].astype(str).tolist()  # Convert to list
    words = []

    for text in text_list:
        doc = nlp(text[:1_000_000])  # Process max 1 million characters per article
        words.extend([token.text.lower() for token in doc if token.is_alpha and token.text.lower() not in stop_words])

    word_counts = Counter(words).most_common(num_words)
    words, counts = zip(*word_counts)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(counts), y=list(words))
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.show()

# Apply the function to the dataset
plot_most_frequent_words(df_train['article'], 'Most Frequent Words in Articles')
plot_most_frequent_words(df_train['highlights'], 'Most Frequent Words in Highlights')
```

```python
import nltk
import spacy
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
from nltk.util import ngrams

# Load spaCy tokenizer (disabling unnecessary components)
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])  
nlp.max_length = 5_000_000  # Increase max length

stop_words = set(stopwords.words('english'))

# Optimized N-gram plotting function
def plot_ngrams(text_series, n=2, num_ngrams=20, title='Most Common N-grams'):
    text_list = text_series[:500].astype(str).tolist()  # Sample first 500 articles
    words = []

    for text in text_list:
        doc = nlp(text[:1_000_000])  # Process in chunks
        words.extend([token.text.lower() for token in doc if token.is_alpha and token.text.lower() not in stop_words])

    n_grams = list(ngrams(words, n))  # Generate n-grams
    freq_dist = Counter(n_grams).most_common(num_ngrams)  # Count frequencies

    ngrams_list, counts = zip(*freq_dist)
    ngrams_list = [' '.join(ngram) for ngram in ngrams_list]

    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(counts), y=list(ngrams_list))
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('N-grams')
    plt.show()

# Apply the function to dataset (Bigrams & Trigrams)
plot_ngrams(df_train['article'], n=2, title='Most Common Bigrams in Articles')
plot_ngrams(df_train['highlights'], n=2, title='Most Common Bigrams in Highlights')
plot_ngrams(df_train['article'], n=3, title='Most Common Trigrams in Articles')
plot_ngrams(df_train['highlights'], n=3, title='Most Common Trigrams in Highlights')
```

```python
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize Sentiment Analyzer (only once)
sid = SentimentIntensityAnalyzer()

# Optimized Sentiment Analysis Function
def plot_sentiment_analysis(text_series, title, sample_size=1000):
    text_sample = text_series[:sample_size].astype(str)  # Use only first 1000 articles
    sentiments = text_sample.apply(lambda x: sid.polarity_scores(x)['compound'])  # Vectorized
    
    plt.figure(figsize=(10, 5))
    sns.histplot(sentiments, bins=50, kde=True)
    plt.title(title)
    plt.xlabel('Sentiment Score')
    plt.ylabel('Frequency')
    plt.show()

# Apply function to dataset
plot_sentiment_analysis(df_train['article'], 'Sentiment Analysis of Articles')
plot_sentiment_analysis(df_train['highlights'], 'Sentiment Analysis of Highlights')
```

```python
import spacy
from collections import Counter

# Load the SpaCy model for NER
nlp = spacy.load("en_core_web_sm")

# Function to plot named entities
def plot_named_entities(text_series, title, batch_size=100):
    entity_counts = Counter()

    for i in range(0, min(len(text_series), 1000), batch_size):
        text_sample = " ".join(text_series[i : i + batch_size])  # Process in batches
        doc = nlp(text_sample)
        entity_counts.update(ent.label_ for ent in doc.ents)

    if entity_counts:
        plt.figure(figsize=(10, 5))
        sns.barplot(x=list(entity_counts.values()), y=list(entity_counts.keys()))
        plt.title(title)
        plt.xlabel('Frequency')
        plt.ylabel('Entity Type')
        plt.show()
    else:
        print(f"No named entities found in the provided text for {title}.")

# Plot named entities in articles and highlights
plot_named_entities(df_train['article'], 'Named Entities in Articles')
plot_named_entities(df_train['highlights'], 'Named Entities in Highlights')
```

```python
# 5. Part-of-Speech (POS) Tagging
import spacy
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Optimized POS Tagging Function
def plot_pos_tags(text_series, title, batch_size=100):
    pos_counts = Counter()

    for i in range(0, min(len(text_series), 1000), batch_size):
        text_batch = " ".join(text_series[i : i + batch_size])  # Process in batches
        doc = nlp(text_batch)  # Process using spaCy
        pos_counts.update(token.pos_ for token in doc)  # Extract POS tags

    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(pos_counts.values()), y=list(pos_counts.keys()))
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('POS Tag')
    plt.show()

# Apply function
plot_pos_tags(df_train['article'], 'POS Tags in Articles')
plot_pos_tags(df_train['highlights'], 'POS Tags in Highlights')
```

```python
# 6. TF-IDF Analysis
from sklearn.feature_extraction.text import TfidfVectorizer
def plot_tfidf_analysis(text, title, num_words=20):
    vectorizer = TfidfVectorizer(max_features=num_words, stop_words='english')
    X = vectorizer.fit_transform(text)
    tfidf_scores = np.mean(X.toarray(), axis=0)
    words = vectorizer.get_feature_names_out()
    sorted_indices = np.argsort(tfidf_scores)[::-1]
    words = [words[i] for i in sorted_indices]
    scores = [tfidf_scores[i] for i in sorted_indices]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=scores, y=words)
    plt.title(title)
    plt.xlabel('TF-IDF Score')
    plt.ylabel('Words')
    plt.show()

plot_tfidf_analysis(df_train['article'], 'TF-IDF Analysis of Articles')
plot_tfidf_analysis(df_train['highlights'], 'TF-IDF Analysis of Highlights')
```
